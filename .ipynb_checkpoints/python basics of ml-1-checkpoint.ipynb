#while loop

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

```python
n = 0
while True:
    if n == 3:
        break
        print("n = ", n)
        n += 1
```

#if -elif(grading)

```python
marks = 75
if marks >= 90:
    print("A")
elif marks >= 80:
    print("B")
elif marks >= 70:
    print("C")
else:
    print("fail")
```

#define a function with parameters and return 

```python
def add (a, b):
    return a + b
```

```python
add(10,20)
```

```python
def stats(nums):
    return min(nums), max(nums), sum(nums)/len(nums)
```

```python
low, high, avg, = stats([10, 20, 30, 40])
```

```python
print(low, high, avg)
```
