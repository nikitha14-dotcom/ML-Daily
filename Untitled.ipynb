```python
messy = pd.DataFrame({
    "Product" : ["Pen","Book","Bag","lamp","Mug"],
    "Price" : [20,np.nan,85,1200,np.nan],
    "rating" :[4.1,4.5,np.nan,4.5,3.9]
})
```

```python
messy
```

```python
messy.isna().sum()
```

```python
filled = messy.copy()
```

```python
filled["Price"]= filled["Price"].fillna(filled["Price"].mean())
```

```python
filled["rating"]= filled["rating"].fillna(filled["rating"].median())
```

```python
print(filled)
```
